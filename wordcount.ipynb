```python
from pyspark import SparkConf
from pyspark import SparkContext
```

```python
conf=SparkConf().setAppName("example1")
sc = SparkContext (conf=conf)
```

```python
rdd=sc.textFile("example.txt")
rdd.count()
```

```python
rdd.collect()
```

```python
words2=rdd.flatMap(lambda line : line.split(" "))
words2.collect()
```

```python
wordCounts = words2.map(lambda word:(word,1))
wordCounts.collect()
```

```python
wordCounts=wordCounts.reduceByKey(lambda a,b:a+b)
wordCounts.collect()
```

```python
wordCounts.saveAsTextFile("C:/example-output")
```
